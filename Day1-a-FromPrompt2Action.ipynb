```python
from dotenv import load_dotenv
import os
from pathlib import Path

env_path = Path('.') / '.env'
load_dotenv(dotenv_path=env_path)

# reference: https://stackoverflow.com/questions/16924471/difference-between-os-getenv-and-os-environ-get
try:
    api_key = os.environ["GOOGLE_API_KEY"]
    os.environ["GOOGLE_GENAI_USE_VERTEXAI"] = "FALSE"
    print("笨 Gemini API key setup complete.")
except Exception as e:
    print(f"沐 Authentication Error: Please make sure you have added 'GOOGLE_API_KEY' in .env file. Details: {e}")
```

```python
from google.adk.agents import Agent
from google.adk.runners import InMemoryRunner
from google.adk.tools import google_search
from google.genai import types

print("笨 ADK components imported successfully.")
```

https://google.github.io/adk-docs/agents/

- name and description: A simple name and description to identify our agent.
- model: The specific LLM that will power the agent's reasoning. We'll use "gemini-2.5-flash-lite".
- instruction: The agent's guiding prompt. This tells the agent its goal is and how to behave.
- tools: A list of tools that the agent can use. To start, we'll give it the google_search tool, which lets it find up-to-date information online.

```python
root_agent = Agent(
    name="helpful_assistant",
    model="gemini-2.5-flash-lite",
    description="A simple agent that can answer general questions using Google Search.",
    instruction="You are a helpful assistant. Use the tools at your disposal to answer user queries" " accurately and concisely.",
    tools=[google_search],
)

print("笨 Root Agent defined.")
```

## Run my agent

```python
runner = InMemoryRunner(agent=root_agent)
print("笨 InMemoryRunner initialized."). # https://google.github.io/adk-docs/runtime/
```

```python
ai_response = await runner.run_debug("WHat is Agent Developemnt Kit from Google? What languages is the SDK available in?" )
```
